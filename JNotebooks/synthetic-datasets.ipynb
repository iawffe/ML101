# Synthetic 2D Datasets

We present here four synthetic datasets that can be used to play using different machine learning algorithms. The implementation is available [here](../Modules/dataset_utils.py). The original code was developed in MATLAB by Jeroen Kools ([link](https://www.mathworks.com/matlabcentral/fileexchange/41459-6-functions-for-generating-artificial-datasets)). 

The goals of this tutorial are:
  - Present synthetic datasets that can be used in future demos.
  - When looking at the datasets try to imagine, which dataset classes' can be distinguished using a linear decision rule (e.g. a line) and which can not.

```python
# Directive to have inline plots opposed to pop-up plots in the notebook 
%matplotlib inline
import numpy as np #Import numerical python with nickname np
import matplotlib.pylab as plt # importing plotting library
import sys
# Adding the Modules folder to the system
MY_UTILS_PATH = '../Modules/'
if not MY_UTILS_PATH in sys.path:
    sys.path.append(MY_UTILS_PATH)
import dataset_utils # loading the synthetic dataset library
```

## Two Spirals dataset

```python
data = dataset_utils.spirals(N = 1000, degrees = 570, start = 90, noise = 0.2)
# Splitting classes 0 and 1 samples into 2 arrays
spiral1 = data[data[:,2] == 0,:-1] 
spiral2 = data[data[:,2] == 1,:-1]

# Plotting the dataset
plt.plot(spiral1[:,0],spiral1[:,1],"x")
plt.plot(spiral2[:,0],spiral2[:,1],"o")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.title("Spirals")
plt.show()
```

## Two Concentric Clusters

```python
data = dataset_utils.concentric_clusters(N = 1000, r1 = 1, r2 = 5, w1 = 0.8,\
                                         w2 = 1.0/3,arms = 64)

# Splitting classes 0 and 1 samples into 2 arrays
cluster1 = data[data[:,2] == 0,:-1]
cluster2 = data[data[:,2] == 1,:-1]

# Plotting the dataset
plt.plot(cluster1[:,0],cluster1[:,1],"x")
plt.plot(cluster2[:,0],cluster2[:,1],"o")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.title("Concentric Clusters")
plt.show()
```

## Four Corners

```python
data = dataset_utils.corners(N = 1000,scale = 0.5,gapwidth = 3,\
                             cornerwidth = 4)

# Splitting classes 0, 1, 2 and 3 into 4 arrays
cluster1 = data[data[:,2] == 0,:-1]
cluster2 = data[data[:,2] == 1,:-1]
cluster3 = data[data[:,2] == 2,:-1]
cluster4 = data[data[:,2] == 3,:-1]

# Plotting the dataset
plt.plot(cluster1[:,0],cluster1[:,1],"x")
plt.plot(cluster2[:,0],cluster2[:,1],"o")
plt.plot(cluster3[:,0],cluster3[:,1],"^")
plt.plot(cluster4[:,0],cluster4[:,1],">")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.title("Four Corners")
plt.show()
```

## Full Moon/ Crescent Moon

```python
data = dataset_utils.crescent_full_moon(N = 1000,r1 = 5,r2 = 10,r3 = 15)

# Splitting classes 0 and 1 samples into 2 arrays
cluster1 = data[data[:,2] == 0,:-1]
cluster2 = data[data[:,2] == 1,:-1]

# Plotting the dataset
plt.plot(cluster1[:,0],cluster1[:,1],"x")
plt.plot(cluster2[:,0],cluster2[:,1],"o")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.title("full Moon/CRescent Moon")
plt.show()
```

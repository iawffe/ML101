# Synthetic 1D Datasets

We present here some common 1D signals. They will be used in future demos related to [Learning the Inverse Discrete 
Fourier Transform](inverse_fourier_transform_1D.ipynb). This demo is optional.

The goals of this tutorial are:
  - Illustrate how to generate simple 1D signals and translated versions of them;
  - Illustrate how to combine 1D signals.

```python
# Display plots within (inline) the notebook
%matplotlib inline 
import matplotlib.pylab as plt # Graphics library
import numpy as np # importing Numpy library with a nickname (np)
```

## Ramp Signal

```python
ramp = np.arange(256)
ramp_samples = np.zeros((30,256))
shift_step = 8
for ii in xrange(30):
    ramp_samples[ii] = np.roll(ramp,shift_step*ii)
    plt.plot(ramp_samples[ii])
plt.show()
np.save("../Data/ramp_samples.npy",ramp_samples)
```

## Sine Signal

```python
x = np.linspace(0,2*np.pi,256)
sinus = np.sin(x)
sin_samples = np.zeros((30,256))
for ii in xrange(30):
    sin_samples[ii] = np.roll(sinus,shift_step*ii)
    plt.plot(sin_samples[ii])
plt.show()
np.save("../Data/sin_samples.npy",sin_samples)
```

## Triangle Signal

```python
triangle = np.zeros(256)
triangle[96:128] = np.arange(32)
triangle[128:159] = np.arange(31)[::-1]
triangle_samples = np.zeros((30,256))
for ii in xrange(30):
    triangle_samples[ii] = np.roll(triangle,shift_step*ii)
    plt.plot(triangle_samples[ii])
plt.show()
np.save("../Data/triangle_samples.npy",triangle_samples)
```

## Exponential Signal

```python
exponencial = np.exp(-np.linspace(0,5,256))
exponencial_samples = np.zeros((30,256))
for ii in xrange(30):
    exponencial_samples[ii] = np.roll(exponencial,shift_step*ii)
    plt.plot(exponencial_samples[ii])
plt.show()
np.save("../Data/exponencial_samples.npy",exponencial_samples)
```

```python
## Sinc Samples
```

```python
x = np.linspace(-4*np.pi,4*np.pi,256)
sinc = np.sin(x)/x
sinc_samples = np.zeros((30,256))
for ii in xrange(30):
    sinc_samples[ii] = np.roll(sinc,shift_step*ii)
    plt.plot(sinc_samples[ii])
plt.show()
np.save("../Data/sinc_samples.npy",sinc_samples)
```

## Step Signal

```python
step = np.zeros(256)
step[116:148] = 1
step_samples = np.zeros((30,256))
for ii in xrange(30):
    step_samples[ii] = np.roll(step,shift_step*ii)
    plt.plot(step_samples[ii])
plt.show()
np.save("../Data/step_samples.npy",step_samples)
```

## Step Mixed with Sinc Signal

```python
indexes = np.arange(30,dtype = int)
np.random.shuffle(indexes)
step_sinc = step_samples[indexes] + sinc_samples[indexes]

for ii in xrange(30):
    plt.plot(step_sinc[ii])
plt.show()
np.save("../Data/step_sinc_samples.npy",step_sinc)
```

## Ramp Mixed with Sine Signal

```python
indexes = np.arange(30,dtype = int)
np.random.shuffle(indexes)
ramp_sinus = ramp_samples[indexes]/255.0 + sin_samples[indexes]

for ii in xrange(30):
    plt.plot(ramp_sinus[ii])
plt.show()
np.save("../Data/ramp_sinus.npy",ramp_sinus)
```

```python
indexes = np.arange(30,dtype = int)
np.random.shuffle(indexes)
exp_sinus = exponencial_samples[indexes] + sin_samples[indexes]

for ii in xrange(30):
    plt.plot(exp_sinus[ii])
plt.show()
np.save("../Data/exp_sinus.npy",exp_sinus)
```

## Ramp Mixed with Triangle Signals

```python
indexes = np.arange(30,dtype = int)
np.random.shuffle(indexes)
ramp_triangle = ramp_samples[indexes] + triangle_samples[indexes]

for ii in xrange(30):
    plt.plot(ramp_triangle[ii])
plt.show()
np.save("../Data/ramp_triangle.npy",ramp_triangle)
```

## Exponential Mixed with Sinc Signals

```python
indexes = np.arange(30,dtype = int)
np.random.shuffle(indexes)
exp_sinc = exponencial_samples[indexes] + sinc_samples[indexes]

for ii in xrange(30):
    plt.plot(exp_sinc[ii])
plt.show()
np.save("../Data/exp_sinc.npy",exp_sinc)
```
